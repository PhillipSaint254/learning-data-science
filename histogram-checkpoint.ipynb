```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
```

```python
plt.hist(blood_sugar)
```

```python
plt.hist(blood_sugar, bins=3)
```

```python
plt.hist(blood_sugar, bins=3, rwidth=0.95)
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95)
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], color="g", rwidth=0.95)
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], color="g", rwidth=0.95, histtype="step")
```

```python
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,150,84,6989,79,120,112,100]
```

```python
plt.xlabel("Sugar Range")
plt.ylabel("No of Patients")
plt.title("Blood Sugar Analysis")
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], color=["green", "orange"], rwidth=0.95, label=["men", "women"])
plt.legend()
```

```python
plt.xlabel("Sugar Range")
plt.ylabel("No of Patients")
plt.title("Blood Sugar Analysis")
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], color=["green", "orange"], rwidth=0.95, label=["men", "women"], orientation="horizontal")
plt.legend()
```
