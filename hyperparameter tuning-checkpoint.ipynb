```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=.3)
```

```python
model = svm.SVC(kernel="rbf", C=30, gamma="auto")
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(svm.SVC(kernel="linear", C=10, gamma="auto"), iris.data, iris.target, cv=5)
```

```python
cross_val_score(svm.SVC(kernel="rbf", C=10, gamma="auto"), iris.data, iris.target, cv=5)
```

```python
cross_val_score(svm.SVC(kernel="rbf", C=20, gamma="auto"), iris.data, iris.target, cv=5)
```

```python
import numpy as np
```

```python
kernels = ["rbf", "linear"]
C = [1,10,20]
average_scores = {}

for kernel in kernels:
    for num in C:
        cv_score = cross_val_score(svm.SVC(kernel=kernel, C=num, gamma="auto"), iris.data, iris.target, cv=5)
        average_scores[kernel + "_" + str(num)] = np.average(cv_score)
        
average_scores
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(svm.SVC(gamma="auto"), {
    'C': [1,10,20],
    "kernel": ["rbf", "linear"]
}, cv=5, return_train_score=False)

clf.fit(iris.data, iris.target)
```

```python
clf.cv_results_
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
model_params = {
    "svm": {
        "model": svm.SVC(gamma="auto"),
        "params": {
            "C": [1,10,20],
            "kernel": ["rbf","linear"]
        }
    },
    "random_forest": {
        "model": RandomForestClassifier(),
        "params": {
            "n_estimators": [1,5,10]
        }
    },
    "logistic_resgression": {
        "model": LogisticRegression(solver="liblinear", multi_class="auto"),
        "params": {
            "C": [1,5,10]
        }
    }
}
```

```python
scores = []

for model_name, mp in model_params.items():
    smh =  GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    smh.fit(iris.data, iris.target)
    scores.append({
        'model': model_name,
        'best_score': smh.best_score_,
        'best_params': smh.best_params_
    })
    
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
```

```python
ex_params = {
    "SVM": {
        "model": svm.SVC(),
        "params": {
            "gamma": ["auto","scale"],
            "C": [1,5,10,15,20],
            "kernel": ["rbf","linear"],
            'degree': [1,2,3,4,5],
            'tol': [1,2,3,4,5],
        }
    },
    "RandomForest": {
        "model": RandomForestClassifier(),
        "params": {
            'n_estimators': [1,5,10,20],
            'random_state': [20,28,42,45,60],
            "min_samples_leaf": [10,20,30,40,50]
        }
    },
    "LinearRegression": {
        "model": LinearRegression(),
        "params": {
            'fit_intercept': [True, False],
            'copy_X': [True, False],
            'positive': [True, False]
        }
    },
    "DecisionTree": {
        "model": DecisionTreeClassifier(max_features=None),
        "params": {
            "random_state": [20,28,42,45,60],
            "max_depth": [1,3,5,8,10,12,15,16,18,20],
            "max_leaf_nodes": [20,32,36,42,48,50,60],
            
        }
    },
    "MultinomialNB": {
        "model": MultinomialNB(),
        "params": {
            'alpha': [0.0001, 0.001, 0.01, 0.1, 0.5, 1.0, 2.0, 10.0],
            'fit_prior': [True, False]
        }
    }
}
```

```python
from sklearn.model_selection import GridSearchCV

scores = []

for mn, mp in ex_params.items():
    ex_model =  GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    ex_model.fit(digits.data, digits.target)
    
    scores.append({
        "model": mn,
        "best_score": ex_model.best_score_,
        "best_params": ex_model.best_params_,
    })
    
pd.DataFrame(scores, columns=["model", "best_score", "best_params"]).head()
```
