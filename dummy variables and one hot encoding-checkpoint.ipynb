```python
import pandas as pd
```

```python
df = pd.read_csv("homeprices (2).csv")
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df, dummies], axis="columns")
merged
```

```python
final = merged.drop(["town","west windsor"], axis="columns")
final
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
X = final.drop("price", axis="columns")
X
```

```python
y = final.price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[1400,0,0]])
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(X,y)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder
```

```python
dfle = df
dfle["town"] = le.fit_transform(dfle['town'], dfle.town)
```

```python
df
```

```python
dfle  # Hot encoded
```

```python
X = df[["town", "area"]].values
X
```

```python
y = dfle.price
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(), [0])
    ],
    remainder='passthrough'  # Pass through the non-categorical columns as is
)

# Fit and transform your data using the ColumnTransformer
ohe = ct.fit_transform(X)
```

```python
ohe
```

```python
X=ohe[:,1:]  # Drop first column
X
```

```python
model.fit(X,y)
```

```python
model.predict([[1,0,2800]])
```

```python
model.predict([[0,1,3400]])
```
